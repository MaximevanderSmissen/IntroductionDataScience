```python
import pandas as pd
```

```python
chessGames = pd.read_csv('Datasets/assignment4.csv', sep=',')
```

```python
chessGames.columns
```

```python
# Ignore all non numerical columns
non_numerical_columns = ['id', 'rated', 'created_at', 'last_move_at', 'victory_status', 'winner', 'increment_code', 'white_id', 'black_id', 'moves', 'opening_eco', 'opening_name']
chessCorrelation = chessGames.drop(non_numerical_columns, axis=1).corr()
chessCorrelation.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

As is visible in the table above, there are few correlations in the given data. 

The strongest positive correlation is 0.63 between black_rating versus white_rating, which is quite plausible as players of a higher rating would want to play against similar higher rated players.

The strongest negative correlation doesn't exist, for all correlations are non-negative.

The weakest correlation is between the turns and the opening_ply (the number of moves in the opening phase)
This could be considered unexpected, for one might assume that the turns would be lower given that the opening ply was short.

```python
# Strongest positive correlation
chessGames.plot.scatter(x='black_rating', y='white_rating', alpha=0.1)
```

```python
# Weakest correlation
chessGames.plot.scatter(x='turns', y='opening_ply', alpha=0.1)
```
